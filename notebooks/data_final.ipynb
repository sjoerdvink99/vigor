```python
import pandas as pd
import numpy as np
import random
import networkx as nx
import os
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from vigor import generate_graphs, encode_categorical, graph_data, plot_attribute
```

```python
%%time
save_data = False
n = 1000
data = {'data': pd.read_csv('../data/generated_graphs_example.csv'), 'rand': graph_data.sample(n)}
# data = {'data': generate_graphs(n), 'rand': graph_data.sample(n)}
for k,v in data.items():
    data[k] = encode_categorical(v, 'graph_type')
if save_data:
    t = int(time.time())
    for k,v in data.items():
        v.to_csv(f'../data/{k}_{t}.csv', index=False)
```

```python
attributes = sorted(list(set([a for b in [v.columns for v in data.values()] for a in b])))
attr_figs = {}
for attr in attributes:
    print(attr)
    fig = plot_attr(data, attr)
    attr_figs[attr] = fig
    plt.show()
```
